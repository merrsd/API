# Срезы датасета по индексам loc | iloc

Срез — способ получения записей из датасета по индексам

### Подгрузка датасета

```python
!gdown --id 1LjVzFpLoB0vfB1wSuc7PxZ0VyiJFNOVS
import pandas as pd
dataset = pd.read_csv('housing_market_dataset.csv')
```

### Кейс 1. Необходимо получить квартиру в датасете под номером 100

 > датасет[название_колонки]



 > датасет.loc[индекс]

```python
dataset.loc[100]
```

### Кейс 2. Необходимо получить часть данных c 1000-ного до 2000-ного индекса

 > датасет.loc[первый_индекс : последний_индекс]

```python
dataset.loc[1000:2000]
```

### Кейс 3. Необходимо получить квартиры по списку индексов.

 > список_индексов = [индекс1, индекс2]

 > датасет.loc[список_индексов]

```python
index_list = [209,1,2,5,10,19,354]
dataset.loc[index_list]
```

### Кейс 4. Кейс в котором индексы не правильные

```python
described_data = dataset.describe()
described_data
```

```python
described_data.loc['max']
```

```python
described_data.loc['min':'max']
```

```python
stat_names = ['min','max'] 0
described_data.loc[stat_names]
```

### Кейс 5. Необходимо собрать топ 100 квартир с наибольшим количеством комнат

```python
sorted_dataset = dataset.sort_values('Количество комнат', ascending=False)
sorted_dataset.head()
```

```python
sorted_dataset.loc[:100]
```

 > датасет.iloc[индекс]

```python
sorted_dataset.iloc[:100]
```

```python
described_data.iloc[:3]
```

### Шаблоны срезов

```python
# Шаблон взятия элемента по индексу
dataset.loc[100]
```

```python
# Шаблон взятия среза по промежутку
dataset.loc[1000:1002]
```

```python
# Шаблон создания нового датасета списку индексов
index_list = [354,1,2]
dataset.loc[index_list].head()
```

```python
# Для не числовых индексов все работает аналогично
describe_data = dataset.describe()
describe_data.loc['max']
```

```python
# Шаблон для взятия среза в хаотично пронумеровнном датасете
sorted_dataset = dataset.sort_values('Количество комнат', ascending=False)
sorted_dataset.iloc[:100]
```
